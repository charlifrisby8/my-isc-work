Q6. Make the plot red with 'x' marking the points.

```python
latitude=ds['latitude']
longitude=ds['longitude']

depth = temperature.sel(latitude=-50.625, longitude=0)
depth.plot(y='depth', color='r', marker='x')
```

 - lat = -50.625, lon = 0
 - lat = 0, lon = 200
 - lat = 0, lon = 320

Hint: use `.sel` to index the lat and lon.

```python

latitude=ds['latitude']
longitude=ds['longitude']

depth1 = temperature.sel(latitude=-50.625, longitude=0) 
depth2 = temperature.sel(latitude=0, longitude=200, method='nearest')
depth3 = temperature.sel(latitude=0, longitude=320, method='nearest')

depth1.plot(y='depth')
depth2.plot(y='depth')
depth3.plot(y='depth')
```
